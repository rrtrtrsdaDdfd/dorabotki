```python
import pandas as pd
df = pd.read_csv('/content/student_data (1).csv')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

target = 'G3'

variables = ['age',	'traveltime',	'studytime',	'famrel',	'freetime']

plt.figure(figsize=(15, 10))

for i, variable in enumerate(variables, 1):
    plt.subplot(3, 2, i)
    sns.kdeplot(data=df, x=variable, hue=target, common_norm=False)
    plt.title(f'{variable} vs {target}')

plt.tight_layout()
plt.show()
```
